```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Mexico_sin_nulos.csv")
data.head()
```

```python
data=data.drop(columns='Unnamed: 0')
data.head()
```

```python
data.info()
```

```python
cualitativas=data.iloc[:,[0,1,2,3,5,6,7,8,9,10,11,12,15,16,17,18,19,20,25,31,38,39,47]]
#float64
cuantitativas=data.iloc[:,[13,14,21,22,23,24,26,27,28,29,30,32,33,34,35,36,37,40,41,42,43,44,45,46,48,49]]
```

```python
fig=plt.figure(figsize=(15,8))
cuantitativas.plot(kind='box',vert=False)
plt.title("valores atipicos del dataframe")
plt.show()
```

**PROCEDIMIENTO DE CUANTILES PATA SUSTITUIR OUTLINERS EN DATAFRAME**

```python
y=cuantitativas
percentile25=y.quantile(0.25)#Q1
percentile75=y.quantile(0.75)#Q3
iqr=percentile75-percentile25
Limite_Superior_iqr=percentile75+1.5*iqr
Limite_Inferior_iqr=percentile25-1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#obtenemos datos limpios del dataframe
data3_iqr=cuantitativas[(y<=Limite_Superior_iqr) & (y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#unmos el dataframe cuantitativo limpio con el dataframe caulitativo
Datos_limpios=pd.concat([cualitativas,data4_iqr],axis=1)
Datos_limpios
```

```python
valores_nulos=Datos_limpios.isnull().sum().sum()
valores_nulos
```

```python
Datos_limpios.to_csv('Datos_limpios_Mexico.csv')
```

```python
fig=plt.figure(figsize=(15,8))
Datos_limpios.plot(kind='box',vert=False)
plt.title("valores atipicos del dataframe")
plt.show()
```

**Procedimiento "DESVIACION ESTANDAR" PARA ELIMINAR OUTLINERS EN DATAFRAME**

```python
#METODO APLICANDO DESVIACION ESTANDAR. ENCUENTRO LOS VALORES
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite Superior permitido", Limite_Superior)
print("Limite Inferior permitido", Limite_Inferior)
```

```python
#obtenemos datos y los outliners se convierten en nulos en el DataFrame 
data3=cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#corroboramos valores atipicos de dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#remplazamos valores atipicos (nulos) del dataframe con mean 
#realizamos una copia del datframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
fig=plt.figure(figsize=(15,8))
data_clean.plot(kind='box',vert=False)
plt.title("valores atipicos del dataframe")
plt.show()
```

```python
data_clean.to_csv('Datos_limpios_Mexico_estandar.csv')
```
